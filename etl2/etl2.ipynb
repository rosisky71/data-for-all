```python
import urllib.request as rq

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0221EN-SkillsNetwork/labs/module%206/Lab%20-%20Extract%20Transform%20Load/data/datasource.zip' 

output_file = 'datasource.zip'

rq.urlretrieve(url, output_file)
```

```python
# Unzip the file
import zipfile
import os

output_folder = 'source'
os.makedirs(output_folder, exist_ok=True)

with zipfile.ZipFile(output_file, 'r') as zip_ref:
    zip_ref.extractall(output_folder)
```

```python
#add utils to path
import sys
project_root = os.path.abspath(os.path.join('..'))
sys.path.append(project_root)


import utils
utils.log_event('extraction process started')
df = utils.extract_data(output_folder) # exract data into one dataframe using utils.extract_data function
utils.log_event('Extraction process completed') 
```

```python
# transform

utils.log_event('Transformation process started, rounding price to 2 decimal places')
df['price'] = round(df['price'], 2) # round price to 2 decimal places
utils.log_event('Transformation process completed')

df
```

```python
#load data to a csv file
utils.log_event('Loading process started, saving transformed data to transformed_data.csv')
utils.load_data('transformed_data.csv', df)
utils.log_event('Loading process completed')
```
