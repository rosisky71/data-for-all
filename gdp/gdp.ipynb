### Objectives

 - Extract required information of countries and gdp from the url.
 - Use Pandas to load and process the tabular data as a dataframe.
 - Use Numpy to manipulate the information contatined in the dataframe.
 - Load the updated dataframe to CSV file.
 


```python
# import libraries
import pandas as pd
import numpy as np
```

```python
URL='https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29'
```

```python
gdp_table = pd.read_html(URL)[3]
gdp_table.describe()
```

```python
# replace table columns headers with numbers
gdp_table.columns = range(0,gdp_table.shape[1])

# retain columns with index 0,2,4,6 (name of country, value of GDP quoted by IMF, value of GDP quoted by World Bank, value of GDP quoted by United Nations) and rows with index 1 to 10 (top 10 countries)
df = gdp_table[[0,2,4,6]].iloc[1:11]

```

```python
# rename columns
df.columns = ['Country', 'GDP 2023', 'GDP 2022', 'GDP 2021']

# export to csv
# df.to_csv('gdp.csv', index=False) 
# 
# #conver gdp of each year to numeric values with astype
df['GDP 2023'] = df['GDP 2023'].astype(int)
df['GDP 2022'] = df['GDP 2022'].astype(int)
df['GDP 2021'] = df['GDP 2021'].astype(int)
# 
```

```python

df['Average GDP'] = np.round(df[['GDP 2023', 'GDP 2022', 'GDP 2021']].mean(axis=1), 2)
```

```python
#export to csv
df.to_csv('gdp.csv', index=False)
```
